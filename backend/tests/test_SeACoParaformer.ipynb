```python
# coding: utf-8
from typing import List, Tuple, TypedDict


class SentenceSpan(TypedDict):
    sentence: str
    start_s: float
    end_s: float


def split_cn_sentences(text: str) -> List[Tuple[int, int, str]]:
    """
    简单中文分句，返回 [(start_idx, end_idx, sentence)]，end_idx 不含。保留句末标点。
    标点集：。！？；（可按需扩展）
    """
    if not text:
        return []
    import re
    spans: List[Tuple[int, int, str]] = []
    start = 0
    for m in re.finditer(r"[。！？；]", text):
        end = m.end()
        sent = text[start:end]
        if sent.strip():
            spans.append((start, end, sent))
        start = end
    if start < len(text):  # 末尾残句
        sent = text[start:]
        if sent.strip():
            spans.append((start, len(text), sent))
    return spans


def map_charpos_to_time_ms(pos: int, text_len: int, seg_ms: List[Tuple[int, int]]) -> int:
    """
    将字符位置 pos (0..text_len) 映射到时间(ms)。
    思路：仅按“有声累计时长”线性分配（忽略静音间隔）。
    seg_ms: [(start_ms, end_ms)] 列表，要求 end_ms > start_ms
    """
    if text_len <= 0 or not seg_ms:
        return 0
    total_voiced = sum(e - s for s, e in seg_ms)
    if total_voiced <= 0:
        return seg_ms[0][0]

    target = total_voiced * (pos / text_len)  # 目标有声累计时长
    acc = 0.0
    for s, e in seg_ms:
        dur = e - s
        if dur <= 0:
            continue
        if acc + dur >= target:
            inside = target - acc
            return int(round(s + inside))
        acc += dur
    return seg_ms[-1][1]


def sentences_with_times(
    text: str,
    timestamps_ms: List[List[int]],
    padding_start_s: float = 0.15,
    padding_end_s: float = 0.25,
) -> List[SentenceSpan]:
    """
    输入：
      - text: 整段转写文本（已带标点）
      - timestamps_ms: 有声片段时间窗（ms），形如 [[s,e], ...]，通常来自 VAD
      - padding_start_s / padding_end_s: 轻微校正起止边界（秒）
    输出：
      - 每句的近似时间范围（秒）：[{sentence, start_s, end_s}, ...]
    """
    seg_ms: List[Tuple[int, int]] = [
        (int(s), int(e)) for s, e in timestamps_ms if isinstance(s, (int, float)) and isinstance(e, (int, float)) and e > s
    ]
    if not text or not seg_ms:
        return []

    sents = split_cn_sentences(text)
    n = len(text)
    rough: List[SentenceSpan] = []
    for s_idx, e_idx, sent in sents:
        st_ms = map_charpos_to_time_ms(s_idx, n, seg_ms)
        ed_ms = map_charpos_to_time_ms(e_idx, n, seg_ms)
        if ed_ms < st_ms:
            ed_ms = st_ms
        rough.append(
            SentenceSpan(
                sentence=sent,
                start_s=round(st_ms / 1000.0, 3),
                end_s=round(ed_ms / 1000.0, 3),
            )
        )

    # 统一做轻微边界校正，避免相邻句重叠与句尾缺字
    adjusted: List[SentenceSpan] = []
    prev_end = 0.0
    for sp in rough:
        st = float(sp["start_s"]) + padding_start_s
        ed = float(sp["end_s"]) + padding_end_s
        if st < prev_end:
            st = prev_end
        if ed <= st:
            ed = st + 0.10
        adjusted.append(
            SentenceSpan(
                sentence=sp["sentence"],
                start_s=round(st, 3),
                end_s=round(ed, 3),
            )
        )
        prev_end = ed
    return adjusted


```

```python
# --- 使用示例 ---
if __name__ == "__main__":
    text = "你好！这是一个测试。我们来验证分句；以及时间对齐？好的。"
    timestamps = [[0, 800], [900, 1600], [2000, 2600]]  # ms

    spans = sentences_with_times(text, timestamps)
    for i, sp in enumerate(spans, 1):
        print(f"{i:02d}. [{sp['start_s']:.3f}s ~ {sp['end_s']:.3f}s] {sp['sentence']}")
```
